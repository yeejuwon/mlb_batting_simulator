```python
# Import datasets
from pathlib import Path

BASE_DIR = Path.cwd()
data_path = BASE_DIR / "data" / "batter_stats_sum.csv"
```

```python
import pandas as pd

b_stat_sum=pd.read_csv(data_path)
print(b_stat_sum)
```

```python
data_path = BASE_DIR / "data" / "fangraphs+platedisc.csv"

fangraphs=pd.read_csv(data_path)
print(fangraphs)
```

```python
# Separate data into left-handed and right-handed batters.
b_stat_sum_L = b_stat_sum[b_stat_sum['stand'] == 'L']
fangraphs_L = fangraphs[fangraphs['Name_FG'].str.endswith('-L') | ~fangraphs['Name_FG'].str.endswith('-R')]

b_stat_sum_R = b_stat_sum[b_stat_sum['stand'] == 'R']
fangraphs_R = fangraphs[fangraphs['Name_FG'].str.endswith('-R') | ~fangraphs['Name_FG'].str.endswith('-L')]

# Merge the datasets for left-handed batters.
merged_df_L = pd.merge(b_stat_sum_L, fangraphs_L, left_on='batter', right_on='key_MLB', how='inner')
merged_df_R = pd.merge(b_stat_sum_R, fangraphs_R, left_on='batter', right_on='key_MLB', how='inner')

# Merge the datasets for right-handed batters.
merged_df = pd.concat([merged_df_L, merged_df_R])

# Combine the left-handed and right-handed merged datasets.
merged_df = merged_df.drop_duplicates()

# Remove duplicate entries and reset the index.
merged_df = merged_df.reset_index(drop=True)

# Display the merged dataset's structure.
print(merged_df)
```

```python
# Display column names for reference.
print(merged_df.columns)
```

```python
# Select key columns for further analysis.
selected_columns = ['Name_FG', 'Name_SC', 'key_FG', 'key_MLB', 'Tm', 'stand',
                    'mean_batspeed', 'std_batspeed', 'mean_swinglength',
                    'std_swinglength', 'attack_angle', 
                    'G', 'PA', 'AB', 'H', '1B', '2B', '3B', 'HR', 'R', 'RBI', 'BB', 'IBB',
                    'SO', 'HBP', 'SF', 'SH', 'GDP', 'SB', 'CS', 'AVG', 'BB%', 'K%', 'BB/K',
                    'OBP', 'SLG', 'OPS', 'ISO', 'BABIP', 'wRC', 'wRAA', 'wOBA', 'wRC+',
                    'GB/FB', 'LD%', 'GB%', 'FB%', 'IFFB%', 'HR/FB', 'IFH%', 'BUH%', 'Pull%',
                     'Cent%', 'Oppo%', 'Soft%', 'Med%', 'Hard%', 'z_swing_percent', 'z_swing_miss_percent',
                     'oz_swing_percent', 'oz_swing_miss_percent', 'oz_contact_percent',
                     'out_zone_swing_miss', 'out_zone_swing', 'out_zone_percent', 'out_zone',
                     'meatball_swing_percent', 'meatball_percent', 'pitch_count_offspeed',
                     'pitch_count_fastball', 'pitch_count_breaking', 'pitch_count',
                     'iz_contact_percent', 'in_zone_swing_miss', 'in_zone_swing',
                     'in_zone_percent', 'in_zone', 'edge_percent', 'edge', 'whiff_percent',
                     'swing_percent', 'f_strike_percent']
merged_sort = merged_df[selected_columns]
print(merged_sort)
```

```python
# Add a new column calculating the ratio of z_swing_percent to oz_swing_percent.
merged_sort['z_swing_pct/oz_swing_pct'] = merged_sort['z_swing_percent'] / merged_sort['oz_swing_percent']

print(merged_sort)
```

```python
# Standardize the format of player names.
merged_sort['Name_FG'] = merged_sort['Name_FG'].str.replace(', L$', '-L', regex=True)
merged_sort['Name_FG'] = merged_sort['Name_FG'].str.replace(', R$', '-R', regex=True)

# Display the final dataset structure and values.
print(merged_sort)
```

```python
# Save the cleaned and processed dataset to a CSV file.
output_path = BASE_DIR / "data" / "cluster_dataset.csv"

merged_sort.to_csv(output_path, index=False)
```
