# Monte Carlo Simulation

```python
# Monte Carlo Simulation Settings
num_innings = 9  # Number of innings to simulate
avg_runs = []  # List to store average runs scored per game

# Get user input for simulation parameters
num_simulations = int(input("How many simulations do you want: "))  # Number of simulations to run
p_throws = input("Pitcher Handedness (R/L): ")  # Pitcher handedness (Right or Left handed)
home_team = input("Home Team: ")  # Name of the home team

# Batting Order
lineup_names = input("Enter lineup as player names (comma-separated): ").split(", ")  # Input lineup as comma-separated player names
lineup_names = [name.strip() for name in lineup_names]  # Remove any extra whitespace

# Map player names to their respective batting clusters
lineup_clusters = []
for name in lineup_names:
    match = player_cluster[player_cluster["Name_FG"] == name]  # Find the player in the dataset
    if not match.empty:
        lineup_clusters.append(match.iloc[0]["batter_cluster"])  # Add the player's cluster if found
    else:
        print(f"Player '{name}' not found in player_cluster.")  # Handle case where player is not found
        lineup_clusters.append(None)  # Append None for players not found

# Display the batting lineup and corresponding clusters
print("Batting Lineup:", lineup_names)
print("Batting Lineup by Swing Clusters:", lineup_clusters)
print()

# Display simulation settings based on pitcher handedness
if p_throws == "R":
    print("Simulating", num_simulations, "Games Against Right Handed Pitcher at", home_team, "...")
else:
    print("Simulating", num_simulations, "Games Against Left Handed Pitcher at", home_team, "...")

# Suggest running more simulations for better accuracy if the number is too low
if num_simulations < 1000:
    print("A higher number of simulations will provide a more accurate answer.")
```

```python
# Adjusting Runs based on Park Factors (2022-2024)

def adjust_runs(runs, home_team, stand):
    # Park factors for left-handed batters (LHB) by team
    lhb_factors = {
        'COL': 1.11, 'BOS': 1.1, 'CIN': 1.05, 'MIA': 1.04, 'PIT': 1.04, 'WSN': 1.04,
        'PHI': 1.03, 'KCR': 1.03, 'MIN': 1.02, 'HOU': 1.01, 'TEX': 1.01, 'BAL': 1.0,
        'ATL': 0.99, 'CHC': 0.99, 'CHW': 0.99, 'CLE': 0.99, 'NYY': 0.99, 'ARI': 0.99,
        'LAA': 0.98, 'MIL': 0.98, 'TOR': 0.98, 'LAD': 0.98, 'STL': 0.98, 'DET': 0.97,
        'SFG': 0.96, 'NYM': 0.96, 'SDP': 0.96, 'TBR': 0.94, 'SEA': 0.91, 'OAK': 0.97
    }

    # Park factors for right-handed batters (RHB) by team
    rhb_factors = {
        'COL': 1.13, 'BOS': 1.04, 'KCR': 1.04, 'CIN': 1.04, 'ARI': 1.03, 'LAA': 1.02,
        'TOR': 1.02, 'MIN': 1.02, 'LAD': 1.01, 'STL': 1.01, 'ATL': 1.01, 'TEX': 1.01,
        'CHW': 1.0, 'MIA': 1.0, 'HOU': 1.0, 'PHI': 1.0, 'NYY': 1.0, 'BAL': 0.99,
        'WSN': 0.99, 'TBR': 0.98, 'PIT': 0.98, 'DET': 0.98, 'NYM': 0.98, 'OAK': 0.97,
        'MIL': 0.97, 'SFG': 0.97, 'SDP': 0.97, 'CHC': 0.96, 'CLE': 0.95, 'SEA': 0.92
    }

    # Determine the appropriate park factor based on handedness
    if stand == 'L':
        factor = lhb_factors.get(home_team, None)  # Get the LHB factor for the home team
    else:
        factor = rhb_factors.get(home_team, None)  # Get the RHB factor for the home team

    # Adjust runs if a valid factor is found
    if factor is not None:
        print(f"Adjusting Runs by {factor} for {stand}HBs at {home_team}")
        return runs * factor  # Multiply runs by the park factor
    else:
        print("Team is Not Found")  # Handle case where team is not found in factors
        return runs

# Example usage: Adjust runs for left-handed batters
runs_lhb = adjust_runs(runs, home_team, 'L')

# Example usage: Adjust runs for right-handed batters
runs_rhb = adjust_runs(runs, home_team, 'R')
```

```python
# Finding the Best Lineup Position for a New Player

# Set probs vs. RHP or probs vs. LHP
probs = probs_rhp if p_throws == 'R' else probs_lhp

def find_best_lineup_position(new_player_name, probs, lineup_names, lineup_clusters, player_cluster, num_simulations, p_throws, runs_lhb, runs_rhb):
    # Check if the new player exists in the player_cluster dataset
    match = player_cluster[player_cluster["Name_FG"] == new_player_name]
    if match.empty:
        print(f"Player '{new_player_name}' not found in player_cluster.")
        return None
    
    # Assign the new player to their corresponding batting cluster
    new_player_cluster = match.iloc[0]["batter_cluster"]
    print(f"New Player '{new_player_name}' assigned to Cluster {new_player_cluster}")
    print("Assigning a higher number of simulations may result in longer processing times.")
    
    # Initialize variables to track the best lineup position and average runs
    best_position = None
    best_avg_runs = -1
    results = []
    
    # Iterate through possible lineup positions
    for i in range(len(lineup_names) + 1):
        # Create a test lineup with the new player inserted at position i
        test_lineup_clusters = lineup_clusters[:i] + [new_player_cluster] + lineup_clusters[i:]
        test_lineup_names = lineup_names[:i] + [new_player_name] + lineup_names[i:]
        
        # Simulate runs for the test lineup
        avg_runs = run_simulation(test_lineup_clusters, probs, num_simulations, p_throws, runs_lhb, runs_rhb)
        results.append((i + 1, avg_runs))
        
        # Update the best position if the current one yields higher average runs
        if avg_runs > best_avg_runs:
            best_position = i + 1
            best_avg_runs = avg_runs
    
    # Print results of the simulations for each lineup position
    print("\nSimulation Results for Each Lineup Position:")
    for position, avg_runs in results:
        print(f"Position {position}: Average Runs = {avg_runs:.4f}")
    
    # Print the best position for the new player
    print(f"\nBest Position for '{new_player_name}': {best_position} (Average Runs: {best_avg_runs:.4f})")
    return best_position


def run_simulation(lineup_clusters, probs, num_simulations, p_throws, runs_lhb, runs_rhb):
    # Initialize a list to store total runs scored in each simulation
    total_runs = []
    for sim in range(num_simulations):
        total_sim_runs = 0 # Track total runs for the current simulation
        lineup_index = 0 # Start at the beginning of the lineup
        
        for _ in range(num_innings):
            current_state = '0000' # Initial base state (empty bases)
            inning_runs = 0 # Track runs scored in the inning
            outs_count = 0 # Track the number of outs in the inning
            
            while outs_count < 3:
                # Get the cluster of the current batter
                batter_cluster = lineup_clusters[lineup_index]
                # Retrieve the probabilities for the batter's cluster
                player_probs = probs[probs['batter_cluster'] == batter_cluster].iloc[0].drop('batter_cluster').values
                cumsum_probs = np.cumsum(player_probs) # Cumulative probabilities for event selection
                event_index = np.searchsorted(cumsum_probs, random.random()) # Select an event
                
                # Update runs based on the batter's handedness and the event
                if batter_cluster in [1, 2, 3, 4]:
                    inning_runs += runs_rhb.loc[current_state].iloc[event_index]
                else:
                    inning_runs += runs_lhb.loc[current_state].iloc[event_index]
                
                # Update outs and base state based on the event
                outs_count += outs.loc[current_state].iloc[event_index]
                current_state = state_transitions.loc[current_state].iloc[event_index]
                
                # Move to the next batter in the lineup
                lineup_index = (lineup_index + 1) % len(lineup_clusters)
            
            # Add inning runs to the total simulation runs
            total_sim_runs += inning_runs
        
        # Append the total runs for this simulation to the list
        total_runs.append(total_sim_runs)
     
     # Return the average runs scored across all simulations
    return np.mean(total_runs)
```

```python
def find_best_cluster_for_position(position, probs, lineup_names, lineup_clusters, num_simulations, p_throws, runs_lhb, runs_rhb, player_cluster):
    # Check if the provided position is valid within the lineup range
    if position < 1 or position > len(lineup_clusters):
        print(f"Invalid position: {position}. Must be between 1 and {len(lineup_clusters)}.")
        return None

    # Notify the user about potential computational costs for higher simulations
    print("Assigning a higher number of simulations may result in longer processing times.")
    
    # Initialize variables to track the best cluster and its corresponding average runs
    best_cluster = None
    best_avg_runs = -1
    results = []

    # Iterate through all unique clusters available in the player_cluster dataset
    for cluster in player_cluster["batter_cluster"].unique():
        # Create a test lineup by replacing the batter at the given position with the current cluster
        test_lineup_clusters = lineup_clusters.copy()
        test_lineup_clusters[position - 1] = cluster
 
        # Run simulations for the test lineup to calculate average runs
        avg_runs = run_simulation(test_lineup_clusters, probs, num_simulations, p_throws, runs_lhb, runs_rhb)
        results.append((cluster, avg_runs))

        # Update the best cluster if the current one yields a higher average run
        if avg_runs > best_avg_runs:
            best_cluster = cluster
            best_avg_runs = avg_runs

    # Sort the results by average runs in descending order for reporting
    sorted_results = sorted(results, key=lambda x: x[1], reverse=True)

    # Print the simulation results for each cluster at the specified position
    print(f"\nSimulation Results for Position {position} against {p_throws}HP (Sorted by Average Runs):")
    for cluster, avg_runs in sorted_results:
        print(f"Cluster {cluster}: Average Runs = {avg_runs:.4f}")
    
    # Print the best cluster and its corresponding average runs
    print(f"\nBest Cluster for Position {position}: {best_cluster} (Average Runs: {best_avg_runs:.4f})")
    return best_cluster
```
