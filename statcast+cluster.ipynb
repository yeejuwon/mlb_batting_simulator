```python
# Import batting_stats_final.csv
from pathlib import Path

BASE_DIR = Path.cwd()
data_path = BASE_DIR / "data" / "batting_stats_final.csv"
```

```python
import pandas as pd

cluster = pd.read_csv(data_path)
print(cluster)
```

```python
# Import Statcast Dataset(Switch Hitter Splitted)
data_path = BASE_DIR / "data" / "statcast_switch_split.csv"

statcast = pd.read_csv(data_path)
print(statcast)
```

```python
# Merge both datasets
merged_df = pd.merge(statcast, cluster[['Name_SC', 'batter_cluster']], left_on='player_name', right_on='Name_SC', how='left')

merged_df = merged_df.dropna(subset=['batter_cluster'])

print(merged_df)
```

```python
# Output 'statcast data + cluster result' dataset
output_path = BASE_DIR / "data" / "statcast+cluster.csv"

merged_df.to_csv(output_path, index=False)
```
